```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

# Reading.
image = np.loadtxt('array_pred.txt')
image=1/image
print(image.shape)
# Display.
print("\033[1m"+ "Depth Map Image: "+ "\033[0m")
fig = plt.figure(dpi=350)
plt.axis('off')
plt.imshow(image)
plt.colorbar(shrink=0.35)
plt.show()


print("\033[1m"+ "Depth Map Image (Gray): "+ "\033[0m")
fig = plt.figure(dpi=350)
plt.axis('off')
plt.imshow(image,cmap = "gray")
plt.colorbar(shrink=0.35)
plt.show()
```
